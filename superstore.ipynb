```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Global_Superstore_5000.csv")
df.head()
```

```python
df.shape        # rows & columns
```

```python
df.info()       # data types
```

```python
df.columns  
```

```python
df.describe() 
```

```python
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
```

```python
df.isnull().sum()
```

```python
df.fillna(0, inplace=True)
```

```python
#Total Sales & Profit
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()

print("Total Sales:", total_sales)
print("Total Profit:", total_profit)
```

```python
#Sales by Category
sales_category = df.groupby('Category')['Sales'].sum()
sales_category
```

```python
#Profit by Region
profit_region = df.groupby('Region')['Profit'].sum()
profit_region
```

```python
#Top 10 Products by Sales
top_products = df.groupby('ProductName')['Sales'].sum().sort_values(ascending=False).head(10)
top_products
```

```python
#Monthly Sales Trend
monthly_sales = df.groupby(df['OrderDate'].dt.to_period('M'))['Sales'].sum()
monthly_sales
```

```python
#Sales by Category (Bar Chart)
sales_category.plot(kind='bar')
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()
```

```python
#rofit by Region (Bar Chart)
profit_region.plot(kind='bar')
plt.title("Profit by Region")
plt.xlabel("Region")
plt.ylabel("Profit")
plt.show()
```
